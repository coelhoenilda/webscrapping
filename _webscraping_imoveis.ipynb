```python
import pandas as pd
from datetime import date
import requests
import time
import random

# Fonte de dados
url_address = 'https://www.olx.com.br/autos-e-pecas/carros-vans-e-utilitarios/gm-chevrolet/opala'
```

```python
# url address acess
url_response = requests.get(url_address)
print(url_response)
```

```python
# user-agent  (avoid site blocking)
headers ={
    'User-Agent': "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.111 Safari/537.36"
}
```

```python
# url acess
url_page= requests.get(url_address, headers=headers)
```

```python
# page bytes to beatifulsoup (methods/functions to scrapping)
from bs4 import BeautifulSoup
url_page = BeautifulSoup(url_page.content, 'html.parser')
```

Automatizando a coleta

```python
df_dados=pd.DataFrame(columns=["ID", "LINK", "TÍTULO", "PREÇO", "LOCAL", "DATA", "INFOS"])
for page in range(1,27):
  # url de cada pagina
  url_address = 'https://www.olx.com.br/autos-e-pecas/carros-vans-e-utilitarios/gm-chevrolet/opala?o='+str(page)
  page_url = requests.get(url_address, headers=headers)
  print (url_address)

  # converter bytes para objeto BeatifulSoap
  page_url = BeautifulSoup(page_url.content, 'html.parser')

  # extrai id
  id = page_url.find_all(class_="sc-12rk7z2-1 huFwya sc-giadOv dXANPZ")
  ids = [x.get_attribute_list("data-lurker_list_id")[0] for x in id]

  # extrai link
  tags_links = page_url.find_all(name = "a", class_="sc-12rk7z2-1 huFwya sc-giadOv dXANPZ")
  links = [x.get_attribute_list("href")[0] for x in tags_links]

  # extrai titulo
  titulo = page_url.find_all(class_="kgl1mq-0 eFXRHn sc-fzsDOv jiCiNE")
  titulo = [x.get_text() for x in titulo]

  # extrai valor
  preco = url_page.find_all(class_="m7nrfa-0 eJCbzj sc-fzsDOv kHeyHD")
  preco = [x.get_text() for x in preco]

  # extrai local do anuncio
  local = url_page.find_all(class_="sc-1c3ysll-1 iDvjkv sc-fzsDOv dTHJIA")
  local = [x.get_text() for x in local]

  # extrai data do anuncio
  data = page_url.find_all(class_="sc-11h4wdr-0 gBAToP sc-fzsDOv dTHJIA")
  data = [x.get_text() for x in data]
  
  # extrai info do anuncio
  infos = url_page.find_all(class_="sc-1ftm7qz-2 ilPFvN")
  infos = [x.get_text() for x in infos]

  # verificando elementos para cada variavevei
  print (len(ids), len(links), len(titulo), len(preco), len(local), len(data), len(infos))
  print (list(zip(ids, links, titulo, preco, local, data, infos)))

  # criar dataframe temp
  dados_tmp = pd.DataFrame(list(zip(ids, links, titulo, preco, local, data, infos)),
                           columns=["ID", "LINK", "TÍTULO", "PREÇO", "LOCAL", "DATA", "INFOS"])
  if len(dados_tmp)>0:
    dados_tmp.loc[:]

    # save links
    dataHora = date.today().strftime('%Y%m%d%H%M')
    dados_tmp.to_csv(f"metadados_olx_{dataHora}{page}.csv", index=True)
  
    df_dados = pd.concat([df_dados,dados_tmp], axis=0,ignore_index=True, keys=ids)

    time.sleep(random.randint(1,5))

    print ("coleta finalizada")


#veiculos_df.to_csv(f"metadados_olx.csv", index=False)
```
